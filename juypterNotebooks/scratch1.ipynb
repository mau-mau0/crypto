```python
import json
import requests
import websockets
```

```python
url = 'https://api.kucoin.com/api/v1/bullet-public'
```

```python
x = requests.post(url)
```

```python
cjson = json.loads(x.text)
```

```python
endpoint = cjson['data']['instanceServers'][0]['endpoint']
token = cjson['data']['token']
```

```python
sub = {
    "id": 1545910660739,
    "type": "subscribe",
    "topic": "/market/ticker:BTC-USDT,ETH-USDT",
    "privateChannel": False,
    "response": True
}
```

```python
async with websockets.connect(endpoint+"?token="+token) as conn:
    await conn.send(json.dumps(sub))
    res = await conn.recv()
    
    while True:
        res = await conn.recv()
        print(res)
    await conn.close()
```

```python
multi = {"id": "1Jpg30DEdU", "type": "openTunnel", "newTunnelId": "bt1", "response": False}
```

```python
multi_eth = {"id": "1JpoPamgFM",
             "type": "subscribe",
             "topic": "/market/ticker:ETH-USDT",
             "tunnelId": "bt1", "response": False}
```

```python
multi_btc = {"id": "1JpoPamgFN",
             "type": "subscribe",
             "topic": "/market/ticker:BTC-USDT",
             "tunnelId": "bt1", "response": False}
```

```python
async with websockets.connect(endpoint+"?token="+token) as conn:
    await conn.send(json.dumps(multi))
    res = await conn.recv()
    
    await conn.send(json.dumps(multi_eth))
    res = await conn.recv()
    await conn.send(json.dumps(multi_btc))
    res = await conn.recv()
    
    counter = 0
    while True and counter < 5:
        res = await conn.recv()
        vals = json.loads(res)
        if float(vals['data']['size']) > 0.5:
            print(res)
            counter += 1
    eth_tunnel = {"id": "1JpoPamgFM", "type": "closeTunnel", "tunnelId": "bt1", "response": False}            
    btc_tunnel = {"id": "1JpoPamgFN", "type": "closeTunnel", "tunnelId": "bt1", "response": False}
    await conn.send(json.dumps(eth_tunnel))
    await conn.send(json.dumps(btc_tunnel))
    await conn.close()
```
